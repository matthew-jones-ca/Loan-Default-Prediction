# Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Data

```python
# Load data from CSV into a DataFrame
df = pd.read_csv('data.csv')

# Show the first 5 rows
df.head()
```

```python
# Remove the 'LoanID' column
df = df.drop(columns='LoanID')

# Relabel the 'Default' column from 0, 1 to 'NoDefault' and 'Default'
df['Default'] = df['Default'].replace({0: 'NonDefault', 1: 'Default'})

# Show the first 5 rows
df.head()
```

```python
# Print the number of observations and variables in the DataFrame
print(f'Number of observations: {df.shape[0]}')
print(f'Number of variables: {df.shape[1]}')
```

```python
# Check for missing values in each variable
df.isnull().sum()
```

```python
# List of categorical variables
categorical_variables = [
    'Education',
    'EmploymentType',
    'MaritalStatus',
    'HasMortgage',
    'HasDependents',
    'LoanPurpose',
    'HasCoSigner',
    'Default'
]

# Output unique labels in each categorical variable
for variable in categorical_variables:
    print(f'{variable}: {df[variable].unique()}')
    print()  # Add a blank line for readability
```
